```python
hisCGLSLAI = pd.read_csv('gs://validation-insitu/histogramCGLSLAI.csv',delimiter=',')
histCGLSLAI = np.empty((0, 100), int)
for row in hisCGLSLAI.prop:
    histCGLSLAI = np.append(histCGLSLAI, np.array([list(eval(row[1:-1]))]),axis=0)
totalsCGLSLAI = np.matlib.repmat((np.sum(histCGLSLAI,axis=0)),100,1)
histCGLSLAI = np.divide(histCGLSLAI,totalsCGLSLAI+10)


hisMODISLAI = pd.read_csv('gs://validation-insitu/histogramLAI.csv',delimiter=',')
histMODISLAI = np.empty((0, 100), int)
for row in hisMODISLAI.prop:
    histMODISLAI = np.append(histMODISLAI, np.array([list(eval(row[1:-1]))]),axis=0)
totalsMODISLAI = np.matlib.repmat((np.sum(histMODISLAI,axis=0)),100,1)
histMODISLAI = np.divide(histMODISLAI,totalsMODISLAI+10)

hisCGLSfAPAR = pd.read_csv('gs://validation-insitu/histogramCGLSfAPAR.csv',delimiter=',')
histCGLSfAPAR = np.empty((0, 100), int)
for row in hisCGLSfAPAR.prop:
    histCGLSfAPAR = np.append(histCGLSfAPAR, np.array([list(eval(row[1:-1]))]),axis=0)
totalsCGLSfAPAR = np.matlib.repmat((np.sum(histCGLSfAPAR,axis=0)),100,1)
histCGLSfAPAR = np.divide(histCGLSfAPAR,totalsCGLSfAPAR+10)

hisMODISfAPAR= pd.read_csv('gs://validation-insitu/histogramfAPAR.csv',delimiter=',')
histMODISfAPAR = np.empty((0, 100), int)
for row in hisMODISfAPAR.prop:
    histMODISfAPAR = np.append(histMODISfAPAR, np.array([list(eval(row[1:-1]))]),axis=0)
totalsMODISfAPAR = np.matlib.repmat((np.sum(histMODISfAPAR,axis=0)),100,1)
histMODISfAPAR = np.divide(histMODISfAPAR,totalsMODISfAPAR+10)


hisCGLSfCOVER = pd.read_csv('gs://validation-insitu/histogramCGLSfCOVER.csv',delimiter=',')
histCGLSfCOVER = np.empty((0, 100), int)
for row in hisCGLSfCOVER.prop:
    histCGLSfCOVER = np.append(histCGLSfCOVER, np.array([list(eval(row[1:-1]))]),axis=0)
totalsCGLSfCOVER  = np.matlib.repmat((np.sum(histCGLSfCOVER,axis=0)),100,1)
histCGLSfCOVER = np.divide(histCGLSfCOVER,totalsCGLSfCOVER +10)


```

```python
fig, ax = plt.subplots(3,1)
fig.set_size_inches(10 ,8)
ax[0].plot(np.linspace(0,10,100),np.log10(totalsMODISLAI[0]+1),':k',label='MODIS')
ax[0].plot(np.linspace(0,10,100),np.log10(totalsCGLSLAI[0]+1),'k',label='CGLS')
ax[0].set_xlabel('SL2P LAI')
ax[0].set_ylabel('log10(#samples)')
ax[0].set_xlim(0,6)
ax[0].legend()


ax[1].plot(np.linspace(0,1,100),np.log10(totalsMODISfAPAR[0]+1),':k',label='MODIS')
ax[1].plot(np.linspace(0,1,100),np.log10(totalsCGLSfAPAR[0]+1),'k',label='CGLS')
ax[1].set_xlabel('SL2P fAPAR')
ax[1].set_ylabel('log10(#samples)')
ax[1].legend()


ax[2].plot(np.linspace(0,1,100),np.log10(totalsCGLSfCOVER[0]+1),'k',label='CGLS')
ax[2].set_xlabel('SL2P fCOVER')
ax[2].set_ylabel('log10(#samples)')
ax[2].legend()
fig.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.1, hspace=0.5)
plt.savefig('sl2pvalpaperfiga.png',bbox_inches='tight')
plt.show()
```
